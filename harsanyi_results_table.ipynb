```python
import numpy as np
import pandas as pd
import os

base_path = 'results/20250508_shapley_optimization_0.025/result/dataset=custom-imdb-for-bertweet-nips2024-ucb-test_model=BERTweet#pretrain_seed=0/players=players-manual_lbl=correct_baseline=unk_bg=ori#/data'

# 初始化结果列表
results = []

# 处理每个sample
for i in range(11):
    sample_path = os.path.join(base_path, f'sample{i}')
    
    # 计算I_and.npy元素和
    i_and = np.load(os.path.join(sample_path, 'I_shapley.npy'))
    interaction_sum = np.sum(i_and)
    
    # 获取rewards_minus_v0.npy最后一个元素
    rewards = np.load(os.path.join(sample_path, 'rewards_minus_v0.npy'))
    last_reward = rewards[-1]
    
    results.append({
        '交互和': interaction_sum,
        '输出': last_reward
    })

# 创建DataFrame
df = pd.DataFrame(results)
df.to_csv('harsanyi_12_results.csv', index=False)
df
```
